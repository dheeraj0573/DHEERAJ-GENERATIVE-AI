```python

# Install necessary libraries (uncomment if not installed)
# !pip install transformers torch --quiet

from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pretrained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Ensure that pad_token_id is set
tokenizer.pad_token = tokenizer.eos_token

# Function to generate text based on a given prompt
def generate_response(prompt, chat_history_ids=None, max_length=150):
    # Encode the user input
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # If there's chat history, concatenate it to the input
    if chat_history_ids is not None:
        input_ids = torch.cat([chat_history_ids, input_ids], dim=-1)

    # Generate a response from the model
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1,
                            pad_token_id=tokenizer.eos_token_id,
                            no_repeat_ngram_size=2,  # Avoids repetition of phrases
                            top_p=0.95,  # Use nucleus sampling to promote diversity
                            top_k=50,  # Consider the top 50 words
                            temperature=0.7,  # Control randomness of predictions
                            do_sample=True)  # Enable sampling to get diverse outputs

    # Decode the output to human-readable text
    response = tokenizer.decode(output[0], skip_special_tokens=True)

    # Return the response and the updated chat history
    return response, output

# Initialize chat history (None at the start)
chat_history_ids = None

print("Chatbot: Hi! I'm your AI chatbot. Type 'exit' to stop chatting.\n")

while True:
    # Take user input
    user_input = input("You: ")

    # Exit loop if user types 'exit'
    if user_input.lower() == 'exit':
        print("Chatbot: Bye! It was nice chatting with you.")
        break

    # Generate a response based on the user input and chat history
    response, chat_history_ids = generate_response(user_input, chat_history_ids)

    # Print the bot's response
    print(f"Chatbot: {response}\n")
```
